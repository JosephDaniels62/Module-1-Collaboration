# Python Unit Testing Assignment

This notebook demonstrates basic unit testing using Python's unittest framework.

```python
# function to test
def add_numbers(a, b):
    return a + b
```

## Writing the Test Cases

```python
import unittest

class TestAddNumbers(unittest.TestCase):

    def test_add_positive_numbers(self):
        self.assertEqual(add_numbers(2, 3), 5)

    def test_add_negative_numbers(self):
        self.assertEqual(add_numbers(-1, -1), -2)

    def test_add_mixed_numbers(self):
        self.assertEqual(add_numbers(-1, 1), 0)
```

## Running the Tests

```python
suite = unittest.TestLoader().loadTestsFromTestCase(TestAddNumbers)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
```

## Test Results Explanation

All tests passing means the function behaves correctly for positive, negative, and mixed inputs.
