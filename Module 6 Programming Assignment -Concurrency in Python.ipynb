```python
## 13.1 Write the current date to a file
```

```python
from datetime import date

today = date.today()
today_string = str(today)

with open("today.txt", "w") as file:
    file.write(today_string)

today_string
```

```python
## 13.2 Read the file into a string
```

```python
with open("today.txt", "r") as file:
    today_string = file.read()

today_string
```

```python
## 13.3 Parse the date
```

```python
from datetime import datetime

parsed_date = datetime.strptime(today_string, "%Y-%m-%d").date()

parsed_date
```

```python
## 15.1 Multiprocessing with random delays
```

```python
import multiprocessing
import time
import random
from datetime import datetime

def worker():
    delay = random.random()
    time.sleep(delay)
    print(f"Process finished at {datetime.now()} after {delay:.3f} seconds")

processes = []

for i in range(3):
    p = multiprocessing.Process(target=worker)
    processes.append(p)
    p.start()

for p in processes:
    p.join()
```
