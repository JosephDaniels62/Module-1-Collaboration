```python
# Demonstrating if/else statements
if number < 5:
    print('The number is less than 5')
elif number == 5:
    print('The number is exactly 5')
else:
    print('The number is greater than 5')
```

```python
# Using a for loop to print list values
my_list = [1, 2, 3, 4, 5]

for value in my_list:
    print(value)
```
