## 13.1 Write the current date to a file
This section writes the current date as a string to a text file named `today.txt`.

```
from datetime import date

today = date.today()
today_string = str(today)

with open("today.txt", "w") as file:
    file.write(today_string)

today_string
```

## 13.2 Read the file into a string
This section reads the contents of `today.txt` into a variable called `today_string`.

```
with open("today.txt", "r") as file:
    today_string = file.read()

today_string
```

## 13.3 Parse the date
This section converts the string from the file back into a date object using `datetime.strptime()`.

```
from datetime import datetime

parsed_date = datetime.strptime(today_string, "%Y-%m-%d").date()

parsed_date
```

## 15.1 Multiprocessing with random delays
This section creates three processes, each waiting a random time and printing the current time.

```
import multiprocessing
import time
import random
from datetime import datetime

def worker():
    delay = random.random()
    time.sleep(delay)
    print(f"Process {multiprocessing.current_process().name} finished at {datetime.now()} after {delay:.3f} seconds")

if __name__ == "__main__":
    processes = []
    for i in range(3):
        p = multiprocessing.Process(target=worker)
        processes.append(p)
        p.start()
    for p in processes:
        p.join()
```
