```python
class Solution:

    def sort012(self, arr):

        low, mid, high = 0, 0, len(arr) - 1

 

        # Traverse through the array

        while mid <= high:

            if arr[mid] == 0:

                # Swap the mid and low elements and move both pointers forward

                arr[low], arr[mid] = arr[mid], arr[low]

                low += 1

                mid += 1

            elif arr[mid] == 1:

                # Move the mid pointer forward

                mid += 1

            else:

                # Swap the mid and high elements and move the high pointer backward

                arr[high], arr[mid] = arr[mid], arr[high]

                high -= 1

       

        return arr
```
